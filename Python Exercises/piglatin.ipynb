```python

#No user input. Simple Pig Latin.
def piglatin (word):
    vowels = ['a', 'e', 'i', 'o', 'u']
    
    firstLetter = word[0]
    otherLetters = word[1:]
    
    if firstLetter in vowels:
        print(word + 'yay')
    else:
        print(otherLetters + firstLetter + 'ay')

piglatin('Computer')
        
```

```python
#Using for loop.
import PyDictionary

from PyDictionary import PyDictionary
    
my_dictionary = PyDictionary()
usr_input = input('Enter a word')
    
firstLetter = usr_input[0]
otherLetters = usr_input[1:]

for i in 'aeiou':
    if i == firstLetter and my_dictionary.meaning(usr_input, True) is not None:
        print(usr_input + 'yay')
        break
    elif i != firstLetter and my_dictionary.meaning(usr_input, True) is not None:
        print(otherLetters + firstLetter + 'ay')
        break
    else:
        print(f'{usr_input} is not in the dictionary!')
        break
```

```python
#Using while statement.
import PyDictionary

from PyDictionary import PyDictionary
    
my_dictionary = PyDictionary()
usr_input = input('Enter a word')
    
firstLetter = usr_input[0]
otherLetters = usr_input[1:]

vowels = ['a', 'e', 'i', 'o', 'u']

while my_dictionary.meaning(usr_input, True) is not None:
        if firstLetter in vowels:
            print(usr_input + 'yay')
            break
        else:
            print(otherLetters + firstLetter + 'ay')
            break
else:
    print(f'{usr_input} is not in the dictionary!')
    
```

```python
#Using while statement. Take one word only.
import PyDictionary

from PyDictionary import PyDictionary
    
my_dictionary = PyDictionary()
usr_input = input('Enter a word')
    
firstLetter = usr_input[0]
secondLetter = usr_input[1]
lastLetter = usr_input[-1]
otherLetters = usr_input[1:]

vowels = ['a', 'e', 'i', 'o', 'u']

while my_dictionary.meaning(usr_input, True) is not None:
        if firstLetter in vowels:
            print(usr_input + 'way')
            break
        if secondLetter and lastLetter in vowels:
            print(usr_input + firstLetter + 'ay')
            break
        else:
            print(otherLetters + firstLetter + 'ay')
            break
else:
    print(f'{usr_input} is not in the dictionary!')
    
```
